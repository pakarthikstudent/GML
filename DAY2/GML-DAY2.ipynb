```python
BFS - graph traversal algorithm 
      starts from root node (or) source node
      visit all the neighbors of current node (before moving to next level of node)
      Queue
      initialize - Queue - stores nodes 
      visited - track the nodes - already visited

create a function - args: graph,start_node
                  - queue - start_node
                  - visited - [] # typecast to set - set all not allows duplicate value
                  - bfs_order = []
                  - all unvisited neighbors of the current node  
                         - use membership neighbor_node in neighbors 
                                 if neighbor not in visited
                                      # add the node to queue # list append operation
```

```python
def f1(graph,node):
    visited = [node]
    queue = [node]
    while(queue):
        node = queue.pop(0)
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)
    return visited
```
